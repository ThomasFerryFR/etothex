```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# machine learning
from sklearn.ensemble import RandomForestRegressor as RFR
```

# Importing the Data

```python
x_raw = pd.read_csv("cleaned_data_v2.csv",encoding = "ISO-8859-1")
y_raw = pd.read_csv("output_data_v2.csv", encoding = "ISO-8859-1")
```

```python
x_raw.head()
```

```python
y_raw.head()
```

```python
y_raw.set_index("Full Name", inplace = True)
x_raw.set_index("Full Name", inplace = True)
```

```python
y = y_raw[['Valuation Increase']].copy()
```

```python
y.head()
```

```python
y.dropna(inplace=True)
```

```python
y=y.convert_objects(convert_numeric=True)
```

```python
y.dropna(inplace=True)
```

```python
y.info()
```

```python
x_raw.head()
```

```python
x=x_raw.copy()
x.drop(['Founding Time','Graduate Degree Time', 'Crunchbase','LinkedIn','Primary Company'],inplace = True, axis=1)
```

```python
x.head()
```

```python
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()
for col in ['Standardized University','Standardized Major', 'Degree Type','Standardized Graduate Institution',"Standardized Graduate Studies",'Graduate Diploma']:
    x[col] = encoder.fit_transform(x[col].fillna('-1'))
```

```python
x.fillna(-1,inplace=True)
```

```python
x.astype(int,inplace=True)
```

```python
df = pd.merge(x,y,left_index=True,right_index=True)
```

```python
df.info()
```

```python
modl = RFR(n_estimators = 1000,n_jobs=-1)

modl.fit(df.drop('Valuation Increase',axis=1),df['Valuation Increase'])

print(modl.score(df.drop('Valuation Increase',axis=1),df['Valuation Increase']))

# Look at importnace of features for random forest

modl.feature_importances_
```

```python
imp = pd.DataFrame( 
        modl.feature_importances_  , 
        columns = [ 'Importance' ] , 
        index = df.drop('Valuation Increase',axis=1)
    )
imp = imp.sort_values( [ 'Importance' ] , ascending = False)
imp
```
