```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    if np.abs(x) < 1:
        return 1
    return 1/np.abs(x)

def g(x):
    if np.abs(x) < 0.11:
        return 1
    else:
        return 0

x = np.arange(-50, 50, 0.1).astype(float)
fy = np.array([f(elem) for elem in x])
gy = np.array([g(elem) for elem in x])
plt.plot(x, fy, label="1 if |x| < 1, else 1/|x|")
plt.plot(x, gy, label="$\delta(x)$\nDirac Delta Function")
plt.xlabel("Difference")
plt.ylabel("Contribution to similarity")
plt.legend()
```
